# *Implementación de las Clases*

```python
# a) Clase Monitor
class Monitor(object):
	"""docstring for Monitor"""
	def __init__(self, marca, tamaño, tipoPantalla, tipoConexion):
		super(Monitor, self).__init__()
		self._marca 		= marca
		self._tamaño 		= tamaño
		self._tipoPantalla 	= tipoPantalla
		self._tipoConexion	= tipoConexion

	# Comandos de consulta
	def getMarca(self):
		return self._marca
	def getTamaño(self):
		return self._tamaño
	def getTipoPantalla(self):
		return self._tipoPantalla
	def getTipoConexion(self):
		return self._tipoConexion

	# Modificacion de atributos
	def setMarca(self, new):
		self._marca = new
	def setTamaño(self, new):
		self._tamaño = new
	def setTipoPantalla(self, new):
		self._tipoPantalla = new
	def setTipoConexion(self, new):
		self._tipoConexion = new

	# Comandos de operaciones
	def __str__(self):
		cantConexiones = len(self._tipoConexion)
		return f'Monitor marca {self._marca} de {self._tamaño}". Pantalla {self._tipoPantalla} con {cantConexiones} conexion/es {self._tipoConexion[:]}.'

# b) Clase Teclado
class Teclado(object):
	"""docstring for Teclado"""
	def __init__(self, marca, idioma, tecladoNumerico, retroiluminacion):
		super(Teclado, self).__init__()
		self._marca 			= marca
		self._idioma 			= idioma
		self._tecladoNumerico 	= tecladoNumerico
		self._retroiluminacion	= retroiluminacion

	# Comandos de consulta
	def getMarca(self):
		return self._marca
	def getIdioma(self):
		return self._idioma
	def getTecladoNumerico(self):
		return self._tecladoNumerico
	def getRetroiluminacion(self):
		return self._retroiluminacion

	# Modificacion de atributos
	def setMarca(self, new):
		self._marca = new
	def setIdioma(self, new):
		self._idioma = new
	def setTecladoNumerico(self, new):
		self._tecladoNumerico = new
	def setRetroiluminacion(self, new):
		self._retroiluminacion = new

	# Comandos operaciones
	def __str__(self):
		if self._tecladoNumerico: # == true
			strNumerico = 'incluye teclado numérico'
		else:
			strNumerico = 'no incluye teclado numérico'

		if self._retroiluminacion: # == true
			strIluminacion = 'tiene retroiluminación'
		else:
			strIluminacion = 'no tiene retroiluminación'

		return f'Teclado marca {self._marca} en idioma {self._idioma}, {strNumerico} y {strIluminacion}'
```

# *Implementación de las Clases*

```python
class Encomienda(object):
    """docstring for Encomienda"""

    estadoEnvio = [
        "Despachado",
        "En viaje",
        "Visita1",
        "Visita2",
        "Recibido",
        "Rechazado",
    ]  # Estado del envio

    def __init__(self, paquete, remitente, destinatario, costo):
        super(Encomienda, self).__init__()
        self._paquete = paquete
        self._remitente = remitente
        self._destinatario = destinatario
        self._costo = costo
        self._idEstado = 0  # Posicion inicial del paquete == Despachado

    def setPaquete(self, nuevo):
        self._paquete = nuevo

    def getPaquete(self):
        return self._paquete

    def setRemitente(self, nuevo):
        self._remitente = nuevo

    def getRemitente(self):
        return self._remitente

    def setDestinatario(self, nuevo):
        self._destinatario = nuevo

    def getDestinatario(self):
        return self._destinatario

    def setCosto(self, nuevo):
        self._costo = nuevo

    def getCosto(self):
        return self._costo

    # def setIdEstado(self, nuevo):
    #     self._idEstado = nuevo

    def getIdEstado(self):
        return self.estadoEnvio[self._idEstado]

    def actualizarEstado(self, codigo=0):
        # 0: proxima visita
        # 1: recibido
        # 2: rechazado
        if len(self.estadoEnvio) - 1 > self._idEstado:
            idVisita1 = self.estadoEnvio.index("Visita1")
            idVisita2 = self.estadoEnvio.index("Visita2")
            idRecibido = self.estadoEnvio.index("Recibido")
            idRechazado = self.estadoEnvio.index("Rechazado")
            if self._idEstado < idVisita1:
                self._idEstado += 1
            elif self._idEstado == idVisita1:
                if codigo == 0:
                    self._idEstado += 1  # pasar a 'visita 2'
                elif codigo == 1:
                    self._idEstado = idRecibido
                elif codigo == 2:
                    self._idEstado = idRechazado
                else:
                    print("Error de codigo")
            elif self._idEstado == idVisita2:
                if codigo == 0:
                    self._idEstado += 1  # pasar a 'Recibido'
                elif codigo == 1:
                    self._idEstado = idRecibido
                elif codigo == 2:
                    self._idEstado = idRechazado
                else:
                    print("Error de codigo")
            else:
                # ya fue recibido o rechazado
                pass

    @classmethod
    def consultarEstadosPosibles(cls):
        return cls.estadoEnvio

    @classmethod
    def agregarEstadoEncomienda(cls, nuevo):
        cls.estadoEnvio.append(nuevo)

    @staticmethod
    # check o cuestiones matematicas


class Paquete(object):
    """docstring for Paquete"""
    def __init__(self, peso, dimensiones):
        self._peso = peso
        self._dimensiones = dimensiones  # lista: alto/ancho/prof


class Persona(object):
    """Persona == Remitente o Destinatario
    clase --> Encomienda( remitente = Persona() )"""

    def __init__(self, nombre, direccion, telefono):
        self.nombre = nombre
        self._direccion = direccion  # objeto Direccion
        self._telefono = telefono

        def setNombre(self, nuevo):
            self._nombre = nuevo

        def getNombre(self):
            return self._nombre

        def setDireccion(self, nuevo):
            self._direccion = nuevo

        def getDireccion(self):
            return self._direccion

        def setTelefono(self, nuevo):
            self._telefono = nuevo

        def getTelefono(self):
            return self._telefono


class Direccion(object):
    """docstring for Direccion
    clase --> Persona( direccion = Direccion() )"""

    def __init__(self, pais, provincia, localidad, calle, numero):
        self._pais = pais
        self._provincia = provincia
        self._localidad = localidad
        self._calle = calle
        self._numero = numero

        def setPais(self, nuevo):
            self._pais = nuevo

        def getPais(self):
            return self._pais

        def setProvincia(self, nuevo):
            self._provincia = nuevo

        def getProvincia(self):
            return self._provincia

        def setLocalidad(self, nuevo):
            self._localidad = nuevo

        def getLocalidad(self):
            return self._localidad

        def setCalle(self, nuevo):
            self._calle = nuevo

        def getCalle(self):
            return self._calle

        def setNumero(self, nuevo):
            self._numero = nuevo

        def getNumero(self):
            return self._numero
```
