````python
# Juan Manuel Roccia - Clases_ej1

# 1) Implementar la clase e incluir los atributos

class Televisor(object):
	"""docstring for Televisor"""
	def __init__(self, marca, totalCanales):
		# super(Televisor, self).__init__()
		self._marca 		= marca
		self._totalCanales 	= totalCanales
		self._canalActual	= 1
		self._volumen		= 25
		self._encendido		= False

		self._muteado = False
		self._ultimoCanala = 1

	# 2) Comandos de consulta 

	def getMarca(self):
		return self._marca

	def getTotalCanales(self):
		return self._totalCanales

	def getCanalActual(self):
		return self._canalActual

	def getVolumen(self):
		if self._muteado:
			return 0
		else:
			return self._volumen	

	def getEncendido(self):
		return self._encendido


	# 3) Comandos modificadores
	
	def setMarca(self, nuevo):
		self._marca = nuevo

	def setTotalCanales(self, nuevo):
		self._totalCanales = nuevo

	def setCanalActual(self, nuevo):
		#if 1<= nuevo <= self._totalCanales:
		if nuevo in range(1, self._totalCanales+1):
				self._ultimoCanal = self._canalActual
				self._canalActual = nuevo


	# 4) Implementar los métodos

	# a) Avanzar al próximo canal 
	def canalSiguiente(self, nuevo):
		self._ultimoCanal = self._canalActual
		self._canalActual += 1
		if self._canalActual > self._totalCanales:
				self._canalActual = 1

	# b) Retroceder al canal anterior
	def canalAnterior(self):
		self._ultimoCanal = self._canalActual
		self._canalActual -= 1
		if self._canalActual < 0:
				self._canalActual = self._totalCanales

	# c) Incrementar el volumen
	def subirVolumen(self):
		self._volumen += 1
		if self._volumen > 100:
			self._volumen = 100
		self._muteado = False

	# d) Bajar el volumen
	def bajarVolumen(self):
		self._volumen -= 1
		if self._volumen < 0 :
			self._volumen = 0
		self._muteado = False

	# e) Silenciar el volumen
	def silenciar(self):
		self._muteado = not self._muteado

	# f) Encender el televisor
	def encender(self):
		self._encendido = True

	# g) Apagar el televisor
	def apagar(self):
		self._encendido = False


	# 6) propuesta para incorporar el método canalPrevio()
	# que permita regresar al canal visto anteriormente
	# Incorporar los atributos y métodos necesarios.
	def canalPrevio(self):
		aux = self._canalActual
		self._canalActual = self._ultimoCanal
		self._ultimoCanal = aux
		# con tuplas
		# (self._canalActual, self._ultimoCanal) = (self._ultimoCanal , self._canalActual)
````

````python
# Juan Manuel Roccia - Clases_ej2

# Clase cliente
class Cliente:
 	"""docstring for Cliente"""
 	def __init__(self, nombre, cuil):
 		self._nombre = nombre
 		self._cuil = cuil

 	# Comandos de consulta
 	def getNombre(self):
 		return self._nombre

 	def getCuil(self):
 		return self._cuil
		
 	# Comandos modificadores
 	def setNombre(self, nuevo):
 		self._nombre = nuevo

 	def setCuil(self, nuevo):
 		self._cuil = nuevo


# Clase Producto
class Producto(object):
	"""docstring for Producto"""
	def __init__(self, nombre, precio):
		self._nombre = nombre
		self._precio = precio

	# Comandos de consulta
	def getNombre(self):
	 	return self._nombre

	def getPrecio(self):
 		return self._precio

 	# Comandos modificadores
	def setNombre(self, nuevo):
 		self._nombre = nuevo

	def setPrecio(self, nuevo):
 		self._precio = nuevo


# Clase Factura
class Factura(object):
	"""docstring for Factura"""
	def __init__(self, cliente, productos):
		self._cliente = cliente
		self._listaProductos = productos
		self._montoTotal = self.calcularMontoTotal()

	# Comandos de consulta
	def getCliente(self):
 		return str(Cliente())
 		
	def getMontoTotal(self):
 		return self._montoTotal
	
	# Comando operacional para el monto total
	def calcularMontoTotal(self):
		suma = 0
		for item in self._listaProductos:
				producto = item[0]
				cant = item[1]
				suma += producto.getPrecio()*cant
		return suma

 	# Comandos de operaciones
	def detallarProductos(self):
		print("Lista de productos")
		for item in self._listaProductos:
				producto = item[0]
				cantidad = item[1]
				print(f" Producto: {producto.getNombre()}, precio unitario: ${producto.getPrecio()} , cantidad: {cantidad}")

	def agregarProductos(self, producto, cantidad = 1):
		for i in range(len(self._listaProductos)):
			prod = self._listaProductos[i][0]
			if prod == producto: # equivalente a prod.equals(producto)
				cant = self._listaProductos[i][1]
				self._listaProductos[i] = (producto, cant + cantidad )
				self._montoTotal = self.calcularMontoTotal()
				return
		# si no hay unidades de ese producto
		self._listaProductos.append((producto, cantidad))
		self._montoTotal = self.calcularMontoTotal()

	def quitarProducto(self, producto):
		for i in range(len(self._listaProductos)):
			prod = self._listaProductos[i][0]
			if prod == producto:
					self._listaProductos.pop(i)
					self._montoTotal = self.calcularMontoTotal()
					return

	def productoMasCaro(self):
		prodMasCaro = self._listaProductos[0][0]
    for item in self._listaProductos:
			producto = item[0]
      if producto.getPrecio() > prodMasCaro.getPrecio():
				prodMasCaro = producto
    return prodMasCaro

  def __iter__(self):
		self._index = 0
		return self

	def __next__(self):
		if self._index < len(self._listaProductos):
			prod = self._listaProductos[self._index][0]
			self._index += 1
				return prod
		else:
				raise StopIteration

	def __contains__(self, item):
		# Retorna True si item esta en la estructura
		# o False en caso contrario
		for x in self._listaProductos:
			prod = x[0]
			if prod == item:
				return True
		return False
````
