# *Implementación de las Clases*

```python
class Programa:
    def __init__(self, nombre, memRequerida):
        self._nombre = nombre
        self._memRequerida = memRequerida
        self._contInstalacion = 0

    # Getters
    def getNombre(self):
        return self._nombre

    def getmemRequerida(self):
        return self._memRequerida

    def getcontInstalacion(self):
        return self._contInstalacion

    # Setters
    def setNombre(self, nuevo):
        self._nombre = nuevo

    def setmemRequerida(self, nuevo):
        self._memRequerida = nuevo

    # Operaciones
    def agregarInstalacion(self):
        self._contInstalacion += 1

    def quitarInstalacion(self):
        self._contInstalacion -= 1


class Celular:
    def __init__(self, modelo, memInterna):
        self._modelo = modelo
        self._memInterna = memInterna
        self._memDisponible = memInterna
        self._programas = []

    # Getters
    def getModelo(self):
        return self._modelo

    def getmemInterna(self):
        return self._memInterna

    def getProgramas(self):
        return self._programas

    def getmemDisponible(self):
        return self._memDisponible

    # Setters
    def setModelo(self, nuevo):
        self._modelo = nuevo

    def setmemInterna(self, nuevo):
        self._memInterna = nuevo
		
	# Operaciones

    def estaInstalado(self, app: Programa):
        if app in self._programas:
            return True
        else:
            return False

    def memoriaSuficiente(self, tamanio):
        if self._memDisponible >= tamanio:
            return True
        else:
            return False

    def instalarPrograma(self, app: Programa):
        info = ""
        if self.estaInstalado(app):
            info = f"La aplicacion ya se encuentra instalada."
        else:
            if self.memoriaSuficiente(app.getmemRequerida()):
                self._programas.append(app)
                app.agregarInstalacion()
                self._memDisponible -= app.getmemRequerida()
                info = f"La aplicacion {app.getNombre()} ha sido instalada exitosamente."
            else:
                info = f"Memoria Insuficiente."
        return info

    def desinstalarPrograma(self, app: Programa):
        if self.estaInstalado(app):
            app.quitarInstalacion()
            self._programas.remove(app)
            self._memDisponible += app.getmemRequerida()
            return True
        else:
            return False

    def agregarMemoria(self, memoria):
        nuevaMemoria = memoria + self._memDisponible
        memoriaOcupada = self._memInterna - self._memDisponible

        if nuevaMemoria >= memoriaOcupada:
             self._memInterna += memoria
             self._memDisponible += memoria
             return True
        else:
             return False
```
