# **Ejercicio 1**
Implementar una clase Vector, que permita almacenar una representación de vectores en dos
dimensiones (x,y). Incorporar los métodos necesarios para que sea posible ejecutar las siguientes
sentencias:

```
a) v1 = Vector(2, 3)
b) v2 = Vector(1, 4)
c) print(v1) # resultado esperado: imprime ‘ Vector = ( x : 2, y : 3) ‘
d) v3 = v1 + v2 # resultado esperado: v3 = Vector(3,7)
e) v4 = 5*v1 # resultado esperado: v4 = Vector(10, 15)
f) v5 = v1.invertir()# resultado esperado: v5 = Vector(-2, -3)
g) for i in v2:
print (i) # resultado esperado: imprime 1, luego imprime 4
h) v6 = v2 - v1 # resultado esperado: v6 = Vector(-1,1)
i) v7 = len(v2) # resultado esperado: 4,1231… (valor aproximado de raíz cuadrada de 17)
j) v8 = v1 + 4 # resultado esperado: Vector(6,7)
k) v9 = 9 + v1 # resultado esperado: Vector(11, 12)
```

````python
# Implementar una clase Vector, que permita almacenar
# una representación de vectores en dos dimensiones (x,y)


class Vector:  # a) y b)
    """docstring for Vector"""

    def __init__(self, x, y):
        self._x = x
        self._y = y

    def getX(self):
        return self._x

    def getY(self):
        return self._y

    def __str__(self):  # c)
        return f"Vector = (x:{self._x}, y:{self._y})"

    # def __add__(self, other):  # d)
    #     return Vector(self._x + other._x, self._y + other._y)

    def __add__(self, other):  # d) y # k)
        if isinstance(other, Vector):
            nuevoX = self._x + other.getX()
            nuevoY = self._y + other.getY()
            return Vector(nuevoX, nuevoY)
        elif isinstance(other, int) or isinstance(other, float):
            nuevoX = self._x + other
            nuevoY = self._y + other
            return Vector(nuevoX, nuevoY)
        else:
            return None

    def __mul__(self, number):  # e)
        return Vector(self._x * number, self._y * number)

    def invertir(self):  # f)
        return Vector(-self._x, -self._y)

    def __iter__(self):  # g)
        yield self._x
        yield self._y

    def __sub__(self, other):  # h)
        return Vector(self._x - other._x, self._y - other._y)

    def __len__(self):  # i)
        return round((self._x**2 + self._y**2) ** 0.5)

    def __radd__(self, other):  # j)
        if isinstance(other, Vector):
            nuevoX = self._x + other.getX()
            nuevoY = self._y + other.getY()
            return Vector(nuevoX, nuevoY)
        elif isinstance(other, int) or isinstance(other, float):
            nuevoX = self._x + other
            nuevoY = self._y + other
            return Vector(nuevoX, nuevoY)
        else:
            return None


# -------- Sentencias --------
# a)
v1 = Vector(2, 3)
# b)
v2 = Vector(1, 4)
# c)
print(v1)  # resultado esperado: 'Vector = (x:2, y:3)'
# d)
v3 = v1 + v2  # resultado esperado: v3 = Vector(3,7)
print(v3)
# e)
v4 = v1 * 5  # resultado esperado: v4 = Vector(10,15)
print(v4)
# f)
v5 = v1.invertir()  # resultado esperado: v5 = Vector(-2, -3)
print(v5)
# g)
for i in v2:
    print(i)  # resultado esperado: imprime 1, luego imprime 4
# h)
v6 = v2 - v1  # resultado esperado: v6 = Vector(-1,1)
print(v6)
# i)
v7 = len(v2)  # resultado esperado: 4,1231...(valor aproximado de raiz cuadrada de 17)
print(v7)
# j)
v8 = v1 + 4  # resultado esperado: Vector(6,7)
print(v8)
# k)
v9 = 9 + v1  # resultado esperado: Vector(11,12)
print(v9)
````

# **Ejercicio 2**

Implementar la clase Sobre, que permita modelar un sobre de papel con sus dimensiones de ancho y
alto. Incorporar los métodos necesarios para que sea posible ejecutar las siguientes sentencias:

```
a) s1 = Sobre(10, 20)
b) s2 = Sobre(8, 15)
c) a = s2 < s1 # resultado esperado: True, ya que s2 podría colocarse dentro de s1
d) b = s1 in s2 # resultado esperado: False, ya que s1 no podría colocarse dentro de s2
e) s3 = s1 / 2 # resultado esperado: [Sobre(10,10), Sobre(10,10)], ya que recorta s1 en dos
sobres más chicos, respetando el ancho
```


````python
# -------- Ejercicio 2 --------


class Sobre(object):
    """docstring for Sobre"""

    def __init__(self, ancho, alto):
        self._ancho = ancho
        self._alto = alto

    def setAncho(self, nuevo):
        self._ancho = nuevo

    def getAncho(self):
        return self._ancho

    def setAlto(self, nuevo):
        self._alto = nuevo

    def getAlto(self):
        return self._alto

    def __lt__(self, otro):  # c)
        return self._alto < otro.getAlto() and self._ancho < otro.getAncho()

    def __iter__(self):
        # logica para recorrer
        # print("Creacion del indice")
        self._idx = 0
        return self

    def __next__(self):
        # print("metodo next con idx = ", self._idx )
        # logica para avanzar y devolver el valor correspondiente
        if self._idx == 0:
            self._idx += 1
            return f" Alto = {self._alto} "
        elif self._idx == 1:
            self._idx += 1
            return f" Ancho = {self._ancho} "
        else:
            raise StopIteration


# -------- Sentencias --------

# a)
s1 = Sobre(10, 20)
# b)
s2 = Sobre(8, 15)
# c)
a = s2 < s1  # resultado esperado True, s2 entraria dentro de s1
print(a)
# d)
b = s1 in s2  # resultado esperado: False s1 no entra en s2
print(b)
# e)
# s3 = s1 / 2  # resultado esperado: [Sobre(10,10), Sobre(10,10)],
# ya que recorta s1 en dos sobres más chicos, respetando el ancho
````
