# ***Ejercicio 1: Relaciones entre clases***

```python
# Juan Manuel Roccia

# 1) Un contacto está definido por un nombre y un teléfono (no es necesario validar, realizar las
# pruebas con valores del tipo correcto). Un contacto es igual a otro cuando sus nombres son
# iguales.
class Contacto(object):
    """docstring for Contacto"""

    def __init__(self, nombre, telefono):
        self.nombre = nombre
        self.telefono = telefono

    def getNombre(self):  # 4)
        return self.nombre

    def getTelefono(self):  # 4)
        return self.telefono

    def setNombre(self, nombre):  # 4)
        self.nombre = nombre

    def setTelefono(self, telefono):  # 4)
        self.telefono = telefono

    def equals(self, contacto):  # 4)
        return self.nombre == contacto.getNombre()


# 2) Una agenda de contactos está formada por un conjunto de contactos (utilizar una
# estructura a elección) y tiene una capacidad máxima que se establece al crearla y no se puede
# modificar posteriormente
class Agenda:
    """docstring for Agenda"""

    def __init__(self, capacidad):
        self.capacidad = capacidad
        self.contactos = []

    # 3) Los métodos de la agenda serán los siguientes:
    def agregarContacto(self, contacto):
        if len(self.contactos) == self.capacidad:
            return False
        else:
            for c in self.contactos:
                if c.getNombre() == contacto.getNombre():
                    return False
        self.contactos.append(contacto)
        return True

    def existeContacto(self, contacto):
        for c in self.contactos:
            if c.getNombre() == contacto.getNombre():
                return True
        return False

    def detallarContactos(self):
        detalle = ""
        for c in self.contactos:
            detalle += c.getNombre() + ": " + c.getTelefono() + "\n"
        return detalle

    def buscarContacto(self, nombre):
        for c in self.contactos:
            if c.getNombre() == nombre:
                return c.getTelefono()
        return "Contacto no encontrado"

    def agendaLlena(self):
        return len(self.contactos) == self.capacidad
```

```python
# Juan Manuel Roccia

# from ejercicio1 import *

# ------- Pruebas 1 -------
contacto1 = Contacto("Juan", "111-111-111")
contacto2 = Contacto("Ana", "222-222-222")
contacto3 = Contacto("Juan", "333-333-333")

print(contacto1.getNombre())  # "Juan"
print(contacto1.getTelefono())  # "111-111-111"
contacto1.setNombre("Juan Pérez")
contacto1.setTelefono("999-999-999")
print(contacto1.getNombre())  # "Juan Pérez"
print(contacto1.getTelefono())  # "999-999-999"
print(contacto1.equals(contacto2))  # False
print(contacto1.equals(contacto3))  # False

# ------- Pruebas 2 -------
agenda = Agenda(5)
contacto1 = Contacto("Juan", "111-111-111")
contacto2 = Contacto("Ana", "222-222-222")
contacto3 = Contacto("Pedro", "333-333-333")
contacto4 = Contacto("Lucia", "444-444-444")
contacto5 = Contacto("Pablo", "555-555-555")
contacto6 = Contacto("Pablo", "666-666-666")

print(agenda.agregarContacto(contacto1))  # True
print(agenda.agregarContacto(contacto2))  # True
print(agenda.agregarContacto(contacto3))  # True
print(agenda.agregarContacto(contacto4))  # True
print(agenda.agregarContacto(contacto5))  # True
print(agenda.agregarContacto(contacto6))  # False

print(agenda.existeContacto(contacto1))  # True
print(agenda.existeContacto(contacto6))  # True

print(
    agenda.detallarContactos()
)  # "Juan: 111-111-111\nAna: 222-222-222\nPedro: 333-333-333\nLucia: 444-444-444\nPablo: 555-555-555"

print(agenda.buscarContacto("Ana"))  # "222-222-222"
print(agenda.buscarContacto("Carlos"))  # "Contacto no encontrado"

print(agenda.agendaLlena())  # True

agenda.agregarContacto(Contacto("Marcos", "777-777-777"))
print(agenda.agendaLlena())  # False
```

# ***Ejercicio 2: Métodos mágicos***

```python
# Juan Manuel Roccia

class Coordenada:
    def __init__(self, x, y):  # Implementacion según el diagrama
        self._x = x
        self._y = y

    # Métodos para las operaciones indicadas en el archivo de prueba.
    def __add__(self, other):
        return Coordenada(self._x + other._x, self._y + other._y)

    def __sub__(self, other):
        if isinstance(other, int):
            return Coordenada(self._x - other, self._y - other)
        else:
            return Coordenada(self._x - other._x, self._y - other._y)

    def __str__(self):
        return f"Coordenada( x = {self._x}, y = {self._y})"


class Trayectoria:
    def __init__(self):  # Implementacion según el diagrama
        self._codigo = ""
        self._listaCoordenadas = []

    def agregarCoordenada(self, coordenada):  # Implementacion según el diagrama
        self._listaCoordenadas.append(coordenada)

    # Métodos para las operaciones indicadas en el archivo de prueba.
    def __gt__(self, other):
        return self.distanciaTotal() > other.distanciaTotal()

    def __iter__(self):
        return iter(self._listaCoordenadas)

    def distanciaTotal(self):
        total = 0
        for i in range(1, len(self._listaCoordenadas)):
            total += self._listaCoordenadas[i]._x - self._listaCoordenadas[i - 1]._x
            total += self._listaCoordenadas[i]._y - self._listaCoordenadas[i - 1]._y
        return total
```

# ***Ejercicio 4: Herencia***

```python
# Juan Manuel Roccia

from abc import ABCMeta, abstractmethod


class Empleado(metaclass=ABCMeta):
    def __init__(self, nombre, salarioBasico, antiguedad=0):
        self._nombre = nombre
        self._salarioBasico = salarioBasico
        self._antiguedad = antiguedad

    def detallarInfo(self):
        return f"Nombre: {self._nombre}, tiene {self._antiguedad} años de antiguedad"

    def calcularSalario(self):
        if self._antiguedad < 5:
            salario = self._salarioBasico + 1000 * self._antiguedad
        elif self._antiguedad < 10:
            salario = self._salarioBasico + 2000 * self._antiguedad
        else:
            salario = self._salarioBasico + 3000 * self._antiguedad
        return salario

    @abstractmethod
    def aumentarSalario(self, incremento):
        pass


# Implementacion de las clases Vendedor y Repartidor
class Vendedor(Empleado):
    # se modifica el método __init__ para inicializar el atributo _ventas
    def __init__(self, nombre, salarioBasico, antiguedad, ventas=0):
        super().__init__(nombre, salarioBasico, antiguedad)
        self._ventas = ventas

    # se sobrescribe el método detallarInfo para incluir la información del atributo _ventas
    def detallarInfo(self):
        return f"Nombre: {self._nombre}, tiene {self._antiguedad} años de antiguedad, realizó {self._ventas} ventas"

    # se sobrescribe el método calcularSalario para agregar un adicional de $100 por cada venta realizada
    def calcularSalario(self):
        salario = super().calcularSalario()
        salario += 100 * self._ventas
        return salario

    # implementación del método aumentarSalario
    def aumentarSalario(self, incremento):
        self._salarioBasico += incremento


class Repartidor(Empleado):
    # se modifica el método __init__ para inicializar el atributo _kmRecorridos
    def __init__(self, nombre, salarioBasico, antiguedad, kmRecorridos=0):
        super().__init__(nombre, salarioBasico, antiguedad)
        self._kmRecorridos = kmRecorridos

    # se sobrescribe el método detallarInfo para incluir la información del atributo _kmRecorridos
    def detallarInfo(self):
        return f"Nombre: {self._nombre}, tiene {self._antiguedad} años de antiguedad, recorrió {self._kmRecorridos} km"

    # se implementa el método realizarEntrega con distintas formas de uso
    def realizarEntrega(self, *args):
        if len(args) == 1:
            # si se recibe un único argumento, se asume que se trata de la cantidad de km recorridos
            self._kmRecorridos += args[0]
        elif len(args) == 2:
            # si se reciben dos argumentos, se asume que el primero es la cantidad de km recorridos y el segundo
            # es el número de entregas realizadas
            self._kmRecorridos += args[0] * args[1]
        else:
            # si se reciben más de dos argumentos, se asume que el primero es la cantidad de km recorridos, el
            # segundo es el número de entregas realizadas y el resto son las zonas en las que se realizaron las
            # entregas
            self._kmRecorridos += args[0] * args[1]
            # se añade un adicional al salario basico por cada zona espe

    # implementación del método aumentarSalario
    def aumentarSalario(self, incremento):
        self._salarioBasico += incremento
```

```python
# Juan Manuel Roccia

# from ejercicio4 import *

vendedor = Vendedor("Juan", 2000, 3, 10)
print(vendedor.detallarInfo())
# se calcula el salario del vendedor
print(vendedor.calcularSalario())
# se aumenta el salario básico del vendedor en $500
vendedor.aumentarSalario(500)
# se vuelve a calcular el salario del vendedor
print(vendedor.calcularSalario())  # debería mostrar 4500

# se crea un repartidor
repartidor = Repartidor("Ana", 1500, 5, 500)
print(repartidor.detallarInfo())
# se calcula el salario del repartidor
print(repartidor.calcularSalario())
# se aumenta el salario básico del repartidor en $300
repartidor.aumentarSalario(300)
# se vuelve a calcular el salario del repartidor
print(repartidor.calcularSalario())
# se utiliza el método realizarEntrega con un único argumento
repartidor.realizarEntrega(100)
print(repartidor.detallarInfo())
# se utiliza el método realizarEntrega con dos argumentos
repartidor.realizarEntrega(10, 5)
print(repartidor.detallarInfo())
# se utiliza el método realizarEntrega con tres argumentos
repartidor.realizarEntrega(20, 3, "zona1", "zona2")
# se imprime la información del repartidor
print(repartidor.detallarInfo())
```
